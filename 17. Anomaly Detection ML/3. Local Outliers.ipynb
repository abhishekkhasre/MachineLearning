```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.datasets import make_circles 

%matplotlib inline
```

```python
X , y = make_circles(n_samples=750 , factor=0.3 , noise=0.1)
```

```python
X
```

```python
plt.scatter(X[:,0] , X[:,1])
```

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
LU = LocalOutlierFactor(n_neighbors=5 , algorithm="auto" )
```

```python
prediction = LU.fit_predict(X)
```

```python
prediction
```

```python
index = np.where(prediction < 0)
```

```python
index
```

```python
plt.scatter(X[:,0] , X[:,1])
plt.scatter(X[index,0], X[index,1] , edgecolors="r")
```
